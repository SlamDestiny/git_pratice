```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
```

```python
normal_d = np.random.normal(10,5,40)
normal_d

#np.random.normalwe chose it mean to be 10 and5 as std.
#The last argument within parentheses shows the numbe rof observations generate by NumPy-40. This is why the the otained one -dimentional arry contantins 40 vaules.
```

```python
#A specific case of the normal distribution is the standard normal distribution. This is a special case when the mean value of distrution is 0 and the #std = 1.

#NumPy provides a tool a generating random number from a standard normal distribution. This time, you need not specify a mean and a std value in #parentheses. Typing a single argument of 40 in parenthese will provide a one-diamonetional arry with 40 elements.

std_norm = np.random.standard_normal(40)
std_norm
```

```python
# If you would like to create a two -dimentional array with standard-normallu distributed values, double the parentheses and designate the dimention in the following way:

std_norm_2 = np.random.standard_normal((4,10))
std_norm_2
```

```python
#Plot the obtained std_norm result on graph

plt.plot(std_norm)
plt.show()
```

```python
range = np.arange(-5,5,0.01)
#The arguments, -5 and 5, designate the end point of the range. The 3rd argument,0.01, specifies the distance between 2 adjacent values.
```
