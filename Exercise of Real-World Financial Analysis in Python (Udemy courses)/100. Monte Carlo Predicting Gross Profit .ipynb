```python
rev_m=170
rev_stdev=20
iterations = 1000
```

```python
rev = np.random.normal(rev_m, rev_stdev, iterations)
rev
```

```python
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show()
```

```python
COGS = - (rev * np.random.normal(0.6,0.1))

plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show()

# COGS is money spent therefore will be addign a negative sign the expression must reflect the mutiplication of revenus by 60%.
# NB will not simulate COGS 1000 times since is already done up. Assign a random COGS value to each of these points.
# COGS is a % of revenus that is why must be mutipily by value extracted from random notmal distribution with a mean of 0,6 and std of 0.1.
# If put 0,6 directly will mean COGS will always = 60% of revenue. % will proberly varia, and std = 0.1 numpy allow us to incorporate the expeccted std.

```
