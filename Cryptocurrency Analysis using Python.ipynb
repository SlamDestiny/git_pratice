```python
# Visualize and graph the crytocurrencies price

my_crypto=df
plt.figure(figsize=(15,12))
#plt.figure(figsize=(12.2,10))
for c in my_crypto.columns.values:
	plt.plot(my_crypto[c], label=c)

plt.title('Cryptocurrency Price')
plt.xlabel('Years')
plt.ylabel('Crypto Price USD ($)')
plt.legend(my_crypto.columns.values, loc='upper left')
plt.show()
```

```python
from sklearn import preprocessing

min_max_scaler=preprocessing.MinMaxScaler(feature_range=(0,100))
scaled=min_max_scaler.fit_transform(df)
scaled
```

```python
# Convert the scaled data into a data frame
df_scale=pd.DataFrame(scaled, columns = df.columns)

# Set the data as index
df_scale=df_scale.set_index(pd.DatetimeIndex(df.index))

# Show the data
pd.set_option('display.max_rows', None)

df_scale
```

```python
my_crypto=df_scale
plt.figure(figsize=(15,12))
#plt.figure(figsize=(12.2,4.5))
for c in my_crypto.columns.values:
	plt.plot(my_crypto[c], label=c)

plt.title('Cryptocurrency Scaled Graph')
plt.xlabel('Years')
plt.ylabel('Individual Crytos Scaled Price from Min to Max')
plt.legend(my_crypto.columns.values, loc='upper left')
plt.show()
```

```python
# Get the daily simple return 
DSR = df.pct_change(1)

# Show the data
DSR
```

```python
# Visualize and graph the crytocurrencies Daily simple return

my_crypto=DSR
plt.figure(figsize=(15,12))
#plt.figure(figsize=(12.2,4.5))
for c in my_crypto.columns.values:
	plt.plot(my_crypto[c], label=c, alpha= .7 ,lw=2)

plt.title('Daily Simple Return')
plt.xlabel('Years')
plt.ylabel('Percentage (in decimal form)')
plt.legend(my_crypto.columns.values, loc='upper left')
```

```python
# Print the crytocurrncies volatility
print('The crypotocurrency volatility:')
DSR.std()
```

```python
 # Print the crytocurrncies simple return
print('The crypotocurrency daily simple return:')
DSR.mean()
```

```python
# Get the correlation matrix

DSR.corr()
```

```python
import seaborn as sns
plt.subplots(figsize=(11,11))
sns.heatmap(DSR.corr(), annot=True, fmt=' .2%')
```

```python
# Get and show the daily cumulative simple returns

DCSR = (DSR+1).cumprod()

# Show the data
DCSR
```

```python
DCSR.describe()
```

```python
my_crypto=DCSR
plt.figure(figsize=(15,12))
#plt.figure(figsize=(12.2,4.5))
for c in my_crypto.columns.values:
	plt.plot(my_crypto[c], label=c, alpha= .7 ,lw=2)

plt.title('Daily Cumaltive Simple Return')
plt.xlabel('Years')
plt.ylabel('Growth of $1 Investment')
plt.legend(my_crypto.columns.values, loc='upper left')
```
